```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#memperlihatkan nilai diatas barplot
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
lima = pd.read_excel("/content/Pengujian Deployment.xlsx",sheet_name="5 cm")
lima.head()
```

```python
sepuluh = pd.read_excel("/content/Pengujian Deployment.xlsx",sheet_name="10 cm")
sepuluh.head()
```

```python
limabelas = pd.read_excel("/content/Pengujian Deployment.xlsx",sheet_name="15 cm")
limabelas.head()
```

```python
plt.figure(figsize=(12,8))
bar_lima=sns.barplot(y="Akurasi", x="Jenis Kelamin", data=lima)
plt.title("Akurasi pada jarak 5 cm", fontsize=30)
plt.rcParams['figure.dpi'] = 300
plt.ylim(1,110)
show_values(bar_lima)
```

```python
plt.figure(figsize=(12,8))
bar_sepuluh=sns.barplot(y="Akurasi", x="Jenis Kelamin", data=sepuluh)
plt.title("Akurasi pada jarak 10 cm", fontsize=30)
plt.rcParams['figure.dpi'] = 300
show_values(bar_sepuluh)
```

```python
plt.figure(figsize=(12,8))
bar_limabelas=sns.barplot(y="Akurasi", x="Jenis Kelamin", data=limabelas)
plt.title("Akurasi pada jarak 15 cm", fontsize=30)
plt.rcParams['figure.dpi'] = 300
show_values(bar_limabelas)
```
